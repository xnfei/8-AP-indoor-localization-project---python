```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
pd.set_option("display.precision", 8)
```

```python
gps =  pd.read_csv('gps.csv')
# get the min and max of (longitude, latitude)
BBox = ((gps.ref_y.min(),   gps.ref_y.max(),      
         gps.ref_x.min(), gps.ref_x.max()))
vertical_one = BBox[0] + (BBox[1] - BBox[0])/3
vertical_two = BBox[1] -(BBox[1] - BBox[0])/3
horizontal_one = BBox[2] + (BBox[3] - BBox[2])/3
horizontal_two = BBox[3] -(BBox[3] - BBox[2])/3
amap = plt.imread('Capture.png')
# the locations of the 8 radio dots
a = np.array([[47.471205,19.062605],
          [47.470997,19.062656],
          [47.470777,19.062807],
          [47.470806,19.063171],
          [47.471282,19.063028],
          [47.471300,19.063312], 
          [47.471118,19.063445],
          [47.470874,19.063478]])
```

```python
fig, ax = plt.subplots(figsize = (8,7))
ax.scatter(gps.ref_y, gps.ref_x, zorder=1, alpha= 0.2, c='b', s=10)
ax.scatter(a[:,1], a[:,0], zorder=1, alpha= 1, c='r', s=100)
ax.set_title('Map')
ax.set_xlim((BBox[0],BBox[1]))
ax.set_ylim((BBox[2],BBox[3]))
ax.vlines([vertical_one, vertical_two], BBox[2], BBox[3], linestyles='dashed', colors='red')
ax.hlines([horizontal_one, horizontal_two], BBox[0], BBox[1], linestyles='dashed', colors='red')
ax.imshow(amap, zorder=0, extent = BBox, aspect= 'auto')
```
